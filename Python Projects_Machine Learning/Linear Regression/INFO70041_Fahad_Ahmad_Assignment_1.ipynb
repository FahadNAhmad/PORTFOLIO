# Evaluating Models: K-fold method

```
#Evaluating models
#K-fold method
#K-Folds Cross validation, split data into k different subsets(or folds)
#Use k-1 subsets to train our data and leave the last subset (or the last fold) as test data
#We then average the model against each of the folds and then finalize our mode.
#After that we test it against the test set.
```

```
from sklearn.model_selection import KFold
kf = KFold(n_splits = 4)
kf.get_n_splits()

X = Crimedata.values[:,:4]
y = Crimedata.values[:,4]

for train_index, test_index in kf.split(X):
    print("TRAIN:", train_index, "\nTEST:", test_index, "\n")
    
    cr_x_train, cr_x_test = X[train_index], X[test_index]
    cr_y_train, cr_y_test = y[train_index], y[test_index]
    
    print("Shapes:\n cr_x_train: {}\n cr_y_train: {}\n cr_x_test: {}\n cr_y_test: {}\n\n".format(cr_x_train.shape, cr_y_train.shape, cr_x_test.shape, cr_y_test.shape))
```

```
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier

cr_x_train, cr_x_test, cr_y_train, cr_y_test = train_test_split(X, y, random_state=4, test_size=0.3)

clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(cr_x_train, cr_y_train) 

y_predict = clf.predict(cr_x_test)

```
