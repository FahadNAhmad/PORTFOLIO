 Data Staging and visualization

```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt, seaborn as sns
```

```python
mall0 = pd.read_csv("Mall_Customers.csv")
mall0.head()
```

```python
mall0.rename({'Annual Income (k$)':'Income', \
              'Spending Score (1-100)':'Spend_score'}, \
             axis=1, inplace=True)
mall0.head()
```

```python
mall0.plot.scatter(x='Income', y='Spend_score', color='gray')
plt.show()
```

 Choosing the Number of Clusters Based on Visual Inspection

```python
mall_scaled = mall0.copy()
cols_to_scale = ['Age', 'Income', 'Spend_score']

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

mall_scaled[cols_to_scale] = scaler.fit_transform\
                             (mall_scaled[cols_to_scale])
```

#### K means clustering - varying number of clusters

```python
from sklearn.cluster import KMeans
```

```python
cluster_cols = ['Income', 'Spend_score']
```

```python
markers = ['x', '*', '.', '|', '_', '1', '2']
```

```python
plt.figure(figsize=[15,10])
for n in range(2,8):
    model = KMeans(n_clusters=n, random_state=42)
    mall_scaled['Cluster']= model.fit_predict\
                            (mall_scaled[cluster_cols])

    plt.subplot(2,3, n-1)
    for clust in range(n):
        temp = mall_scaled[mall_scaled.Cluster == clust]
        plt.scatter(temp.Income, temp.Spend_score, \
                    marker=markers[clust], \
                    label="Cluster "+str(clust), \
                    color='gray')
        plt.title("N clusters: "+str(n))
        plt.xlabel('Income')
        plt.ylabel('Spend_score')
        plt.legend()

plt.show()
```

```python
K = 3
```

```python
model = KMeans(n_clusters=K, random_state=42)
model.fit(mall_scaled[cluster_cols])
```

```python
X = mall_scaled[cluster_cols]
```

```python
inertia_scores = []
for K in range(2,11):
    inertia = KMeans(n_clusters=K, random_state=42).fit(X)\
                                                   .inertia_
    inertia_scores.append(inertia)
```

 Clustering Data Using the k-prototypes Method

```python
import pandas as pd
bank0 = pd.read_csv("Bank_Personal_Loan_Modelling-2.csv")
bank0.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
bank_scaled = bank0.copy()
bank_scaled['Income'] = scaler.fit_transform(bank0[['Income']])
```

Applying k-prototypes

```python
from kmodes.kprototypes import KPrototypes
```

```python
cluster_cols = ['Income', 'Education']
```

```python
kp = KPrototypes(n_clusters=3, random_state=42)
bank_scaled['Cluster'] = kp.fit_predict\
                         (bank_scaled[cluster_cols],\
                          categorical=[1])
```

```python
res = bank_scaled.groupby('Cluster')['Education']\
                 .value_counts(normalize=True)

res.unstack().plot.barh(figsize=[9,6], \
                        color=['black','lightgray','dimgray'])
plt.show()
```

Using Silhouette Score to Pick Optimal Number of Clusters

```python
cluster_cols = ['Income', 'Spend_score']
X = mall_scaled[cluster_cols]
```

```python
model = KMeans(n_clusters=3, random_state=42)
cluster_assignments = model.fit_predict(X)
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(X, cluster_assignments)
print(silhouette_avg)
```

```python
silhouette_scores = []

for K in range(2, 11):
    model = KMeans(n_clusters=K, random_state=42)
    cluster_assignments = model.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_assignments)
    silhouette_scores.append(silhouette_avg)
```

```python
plt.figure(figsize=[7,5])
plt.plot(range(2,11), silhouette_scores, color='gray')
plt.xlabel("Number of clusters: K")
plt.ylabel('Avg. Silhouette Score')
plt.show()
```
