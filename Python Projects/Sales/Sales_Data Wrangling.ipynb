```python
import pandas as pd
```

```python
sales = pd.read_csv("sales.csv")
sales.head()
```

```python
sales.groupby('Order method type').sum()\
     .plot(kind = 'bar', y = 'Revenue', color='gray')
```

```python
sales.groupby('Year')[['Revenue', 'Planned revenue',\
                       'Gross profit']].plot(kind= 'box',\
                                             color='gray')
```
