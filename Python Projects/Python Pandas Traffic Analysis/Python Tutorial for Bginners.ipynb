# 6. While Loops and Breaks

```python
total = 0
for i in range(1, 5):
    total += i
print(total)
```

```python
total2 = 0
j = 1
while j < 5:
    total2 += j
    j += 1
print(total2)
```

```python
given_list = [5, 4, 4, 3, 1]
total3 = 0
i = 0
while i < len(given_list) and given_list[i] > 0:
    total3 += given_list[i]
    i += 1
print(total3)
```

```python
given_list2 = [5, 4, 4, 3, 1, -2, -3, -5]
total4 = 0
for element in given_list2:
    if element <= 0:
        break
    total4 += element
print(total4)
```

# 7. More For Loops

```python
a = ["apple", "banana", "republic"]
for element in a:
    print(element)
```

```python
for i in range(len(a)): # 0, 1, 2
    for j in range(i + 1):
        # i = 0 -> j = 0
        # i = 1 -> j = 0, 1
        # i = 2 -> j = 0, 1, 2
        print(a[i])
```

```python
# compute the sum of all multiples of 3 and 5 that are less than 100?
print(list(range(1, 100)))
```

```python
total = 0
for i in range(1, 100):
    if i % 3 == 0 or i % 5 == 0:
        total += i
print(total)
```

```python
# sum of negatives
given_list = [7, 5, 4, 4, 3, 1, -2, -3, -5, -7]

total2 = 0
j = len(given_list) - 1
while given_list[j] < 0:
    total2 += given_list[j]
    j -= 1
print(total2)
```

# 14. CodingBat

```python
# sleep_in
def sleep_in(weekday, vacation):
    return not weekday or vacation
```

```python
# string_times
def string_times(str, n):
    return str * n
string_times('hi', 4)
```

```python
# hello_name
def hello_name(name):
    return 'Hello ' + name + '!'
hello_name('Bob')
```

```python
# first_last6
def first_last6(nums):
    return nums[0] == 6 or nums[-1] == 6
first_last6([1, 2, 6])
```

```python
# double_char
def double_char(str):
    to_return = ''
    for c in str:
        to_return += c*2
    return to_return
double_char('The')
```

```python
# count_evens
def count_evens(nums):
    count = 0
    for x in nums:
        if x % 2 == 0:
            count += 1
    return count
count_evens([2, 1, 2, 3, 4])
```
