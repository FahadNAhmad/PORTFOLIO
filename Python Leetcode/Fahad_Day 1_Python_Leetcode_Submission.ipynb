# Q53: Maximum Subarray

Approach 1:

For the first number, find all the combinations of contiguous sub-array and check the sum.
When the new sub-arrary sum is greater than the global sum, then replace the global sum.
Repeat this step onto the rest of the list

```python
nums = [-2,1,-3,4,-1,2,1,-5,4]
```

```python
def maxSubArray_2(nums):
  #initially maximum sum is set as zero
  maxsum = 0
  #Loop through list starts with -2
  for i in range(len(nums)):
     #current sum set to 0
     currsum = 0
     #first loop, -2+1, next loop -2+1+-3
     for j in range(i+1, len(nums)):
        #e.g., -2+1 = -1
        currsum = currsum + nums[j]
        #if current sum is greater than maxsum
        if currsum > maxsum:
           maxsum = currsum
  return (maxsum)
     
     
maxSubArray_2(nums)
```

```python
#An easier way, by storing the current sum and comparing to maxsum, if greater
#then maxsum becomes the temporary sum
#using the sum(nums[i:k]) notation
def maxSubArray_2(nums):
    max_sum = 0
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            temp_sum = sum(nums[i:j + 1])
            if temp_sum > max_sum:
                max_sum = temp_sum
    return max_sum

maxSubArray_2(nums)
```

# Q136. Single Number

Approach 2:

Create a dictionary to store each value and its associated appearances.
If the value is seen for the first time, add to the dictionary and set counter as 1.
Otherwise, add 1 to the counter.
Select the value where counter equals to 1

```python
nums = [4,1,2,1,2]
```

```python
nums = [4,1,2,1,2]
def singleNumber_2(nums):
  dic = {}
  for num in nums:
      if (num in dic):
      #add 1 to the value for the key in dictionary, like a frequency
         dic[num] += 1
      else:
      #if value not in dictionary, set 1 as value for the key or number
         dic[num] = 1

  #Loop through the dictionary of keys and values, only return key with value of 1
  for key, value in dic.items():
      if value == 1:
         return key

singleNumber_2(nums)
```

```python
nums = [4,1,2,1,2]
def singleNumber_2(nums):
  dic = {}
  for num in nums:
      dic[num] = dic.get(num, 0) + 1
  for key, value in dic.items():
      if value == 1:
         return key

singleNumber_2(nums)
```
