```r
IntentionalYesNo <- aggregate(Correct ~ SID + display + TestCondition, dataIntYesNo, mean)
```

```r
IntentionalYesNo
```

```r
factor(IntentionalYesNo$display, levels = c("Picture Test Trial", "Sound Test Trial"), labels=c("Picture","Sound"))
```

```r
#Need to group in appropriate format to run Anova in R
IntentionalYesNo<-IntentionalYesNo %>% group_by(SID, display, TestCondition) %>% summarize(MeanCorrect = mean(Correct))
```

```r
IntentionalYesNo
```

```r
if(!require(psych)){install.packages("psych")}
if(!require(FSA)){install.packages("FSA")}
if(!require(car)){install.packages("car")}
if(!require(multcompView)){install.packages("multcompView")}
if(!require(lsmeans)){install.packages("lsmeans")}
if(!require(Rmisc)){install.packages("Rmisc")}
if(!require(ggplot2)){install.packages("ggplot2")}
if(!require(rcompanion)){install.packages("rcompanion")}
```

```r
library(FSA)

Summarize(MeanCorrect ~ TestCondition + display, data=IntentionalYesNo, digits=3)
```

```r
model = lm(MeanCorrect ~ TestCondition + display + TestCondition:display, data = IntentionalYesNo)
summary(model)
```

```r

library(car)

Anova(model, type = "II")
```
