Step 1: Create two column variables for CorrecIncorr and CorrIncorrRT then Shift the second row up by one with the function "lead"

Step 2:Create two column variables for ConfR and CfRT then shift the third row up by one with the function "lead"

```r
dataIntYesNo <- dataIntYesNo %>% filter(ZoneType %in% c('response_keyboard', 'response_rating_scale_likert_active')) %>% mutate(CorrectIncorr = Correct, CorrIncorrRT = ReactionTime) %>% mutate_at(c("CorrectIncorr", "CorrIncorrRT"), list(lead), n = 1) %>% mutate(ConfR = Response, CfRT = ReactionTime) %>% mutate_at(c("ConfR", "CfRT"), list(lead), n = 2)  
```
