# Repeated Measures ANOVA

```r
str(Intentionalcond)
```

```r
factor(Intentionalcond$display, levels = c("Picture Test Trial", "Sound Test Trial"), labels=c("Picture","Sound"))
```

```r
factor(Intentionalcond$TestCondition, levels = c("Exemplar", "Novel"))
```

```r
str(Intentionalcond)
```

```r
Intentionalcond$display <- as.factor(Intentionalcond$display)
Intentionalcond$TestCondition<-as.factor(Intentionalcond$TestCondition)
```

```r
str(Intentionalcond)
```

```r
Intentionalcond
```

```r
Intentionalcond$Incorrect<-1-Intentionalcond$Correct
```

```r
Intentionalcond
```

```r
#Need to group in appropriate format to run Anova in R
Intentionalcond<-Intentionalcond %>% group_by(SID, display, TestCondition) %>% summarize(MeanCorrect = mean(Correct))
```

```r
Intentionalcond
```

```r
if(!require(psych)){install.packages("psych")}
if(!require(FSA)){install.packages("FSA")}
if(!require(car)){install.packages("car")}
if(!require(multcompView)){install.packages("multcompView")}
if(!require(lsmeans)){install.packages("lsmeans")}
if(!require(Rmisc)){install.packages("Rmisc")}
if(!require(ggplot2)){install.packages("ggplot2")}
if(!require(rcompanion)){install.packages("rcompanion")}
```

```r
library(FSA)

Summarize(MeanCorrect ~ TestCondition + display, data=Intentionalcond, digits=3)

```

```r
model = lm(MeanCorrect ~ TestCondition + display + TestCondition:display, data = Intentionalcond)
summary(model)
```

```r

library(car)

Anova(model, type = "II")
```

```r
#library(tidyverse)
#install.packages(c("ggplot2", "ggpubr", "tidyverse", "broom", "AICcmodavg"))
#library(ggplot2)
#library(ggpubr)
#library(tidyverse)
#library(broom)
#library(AICcmodavg)
#install.packages("rstatix")
#library(rstatix)

res.aov <- aov(MeanCorrect ~ TestCondition*display, data= Intentionalcond)
summary(res.aov)
```

```r
tukey.two.way<-TukeyHSD(res.aov)

tukey.two.way
```
