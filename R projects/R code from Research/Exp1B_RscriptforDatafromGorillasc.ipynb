```r
library(FSA)

Summarize(Correct ~ TestCondition + display, data=IntentionalYesNo, digits=3)
```

```r
model = lm(Correct ~ TestCondition + display + TestCondition:display, data = IntentionalYesNo)
summary(model)
```

```r

library(car)

Anova(model, type = "II")
```
