```r
#Create corrected Exemplar, ExemplarIncorr, Novel, NovelIncorr; based on the formula
dprimetable1 <- dprimetable1 %>% mutate(Exemplar= if_else(Exemplar == '1', (13-0.5)/13,Exemplar)) %>% mutate(Novel= if_else(Novel == '1', (13-0.5)/13,Novel)) %>% mutate(ExemplarIncorr = if_else(ExemplarIncorr == '0', 0.5/13,ExemplarIncorr)) %>% mutate(NovelIncorr= if_else(NovelIncorr == '0', 0.5/13,NovelIncorr))
dprimetable1
```

```r
dprimetable2<-dprimetable1 %>% mutate(ExemplarHRFAR = Exemplar - ExemplarIncorr)  %>% mutate(NovelHRFAR = Novel -NovelIncorr) %>% mutate(dprimeTargetExemplar = (qnorm(Exemplar)-qnorm(ExemplarIncorr))/sqrt(2)) %>% mutate(dprimeTargetNovel = (qnorm(Novel)-qnorm(NovelIncorr))/sqrt(2)) 
dprimetable2
```

```r
dprime2avg <- dprimetable2 %>% group_by(display)%>% summarise(mean_NovelHRFAR = mean(NovelHRFAR),mean_ExemplarHRFAR = mean(ExemplarHRFAR), mean_dprimeTargetExemplar=mean(dprimeTargetExemplar), mean_dprimeTargetNovel=mean(dprimeTargetNovel))
dprime2avg
```
