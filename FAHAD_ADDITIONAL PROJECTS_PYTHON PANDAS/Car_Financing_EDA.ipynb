Renaming and Deleting Columns
Rename columns
Two ways to rename dataframe columns
1.Dictionary substitution: useful if want to rename a few of the columns
2.List replacement: requires a full list of names 

```python
#DataFrame before renaming columns
df.head()
```

```python
#This wont work as there is a space in the column name
#I want to fix that
df['Principal Paid']
```

## Removing or Filling in Missing Data
This is an important subject as before you can graph data, you should make sure you aren't trying to graph some missing values as that can cause an error or misinterpretation of the data. 

```python
df.info()
```

### Remove Missing Values
You can remove missing values by using the `dropna` method. 

```python
# You can drop entire rows if they contain 'any' nans in them or 'all'
# this may not be the best strategy for our dataset
df[30:40].dropna(how = 'any')
```

### Filling in Missing Values
There are a [variety of ways to fill in missing values](https://pandas.pydata.org/pandas-docs/stable/user_guide/missing_data.html). 

```python
# Looking at where missing data is located
df['interest_paid'][30:40]
```

```python
# Filling in the nan with a zero is probably a bad idea. 
df['interest_paid'][30:40].fillna(0)
```

```python
# back fill in value
df['interest_paid'][30:40].fillna(method='bfill')
```

```python
# forward fill in value
df['interest_paid'][30:40].fillna(method='ffill')
```

```python
# linear interpolation (filling in of values)
df['interest_paid'][30:40].interpolate(method = 'linear')
```

```python
# Interest paid before filling in the nan with a value
df['interest_paid'].sum()
```

```python
# Fill in with the actual value
interest_missing = df['interest_paid'].isna()
df.loc[interest_missing,'interest_paid'] = 93.24
```

```python
# Interest paid after filling in the nan with a value
df['interest_paid'].sum()
```

```python
# Notice we dont have NaN values in the DataFrame anymore
df.info()
```

Matplotlib:A powerful tool capable of porducing a complex publication-quality of figures with fine layout control in two or three dimesions.

```python
#The inline flag will use the appropriate backend to make figures appear inline in the notebook
%matplotlib inline

import pandas as pd
import numpy as np

#plot is an alias for the matplotlib.pyplot' module
import matplotlib.pyplot as plt

#import seaborn library (wrapper of matplotlib)
import seaborn as sns
```

```python
# Load car loan data into a pandas datafrme from a csv file
df = pd.read_csv('https://raw.githubusercontent.com/FahadNAhmad/DATASETS/master/table_i702t60.csv')
```

```python
#Checking to make sure we dont have nans in our dataframe
#It is not easy to directly plot data that contains nans
df.info()
```

```python
#The values attribute converts a column of values into a numpy array
month_number = df.loc[:, 'month'].values
interest_paid = df.loc[:, 'interest_paid'].values
principal_paid = df.loc[:, 'principal_paid'].values
```

Choose Figure Style
We will use plt.style.available to select an appropriate aesthetic styles for our figures. The default style is not the most aesthetically pleasing.

```python
#Our style choices
plt.style.available
```

```python
#We are currently using default plot
#So this will not channge much our plot
plt.style.use('classic')
plt.plot(month_number, interest_paid)
plt.plot(month_number, principal_paid)
```

```python
plt.style.use('fivethirtyeight')
plt.plot(month_number, interest_paid)
plt.plot(month_number, principal_paid)
```

```python
plt.style.use('seaborn')
plt.plot(month_number, interest_paid)
plt.plot(month_number, principal_paid)
```

```python
plt.style.use('seaborn')
plt.plot(month_number, interest_paid, marker = '.', markersize = 10)
plt.plot(month_number, principal_paid, marker = '.', markersize = 10)
```

```python
#Using hex strings
# '#00000' is black
# '#000FF' is blue
plt.plot(month_number, interest_paid, c ='#000000', marker = '.', markersize = 10)
plt.plot(month_number, principal_paid, c ='#0000FF', marker = '.', markersize = 10)
```

```python
#Want one plot
fig, axes = plt.subplots(nrows = 1, ncols = 1);
#k for the color blue
axes.plot(month_number, interest_paid, c = 'k');
#b for the color black
axes.plot(month_number, principal_paid, c = 'b');
axes.set_xlabel('Month', fontsize = 22);
axes.set_ylabel('Dollars', fontsize = 22);
axes.set_title('Interest and Principal Paid Each Month', fontsize = 22)
```

```python
#Add grid lines
#Want one plot
fig, axes = plt.subplots(nrows = 1, ncols = 1);
#k for the color blue
axes.plot(month_number, interest_paid, c = 'k', label = 'Interest');
#b for the color black
axes.plot(month_number, principal_paid, c = 'b', label = 'Principal');
axes.grid(axis = 'x')
axes.set_xlabel('Month', fontsize = 22);
axes.set_ylabel('Dollars', fontsize = 22);
axes.set_title('Interest and Principal Paid Each Month', fontsize = 22)
axes.legend(loc = "center right")
```
